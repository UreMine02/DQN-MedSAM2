```python
from PIL import Image
import matplotlib.pyplot as plt
from glob import glob
import os
```

```python
rl_dir = "output/msd_task02+grpo+icl/2026-02-11-10-07-12/vis"
fifo_dir = "output/msd_task02+grpo+icl+freeze/2026-02-11-14-41-19/vis"
```

```python
def key(p):
    parts = p.split("_")
    vol = int(parts[-4].split(".")[0])
    obj = int(parts[-3])
    frame = int(parts[-2][3:])

    return (vol, obj, frame)
    
rl_preds = sorted(glob(os.path.join(rl_dir, "*")), key=key)
fifo_preds = sorted(glob(os.path.join(fifo_dir, "*")), key=key)
```

```python
idx = 142
print(rl_preds[idx])
print(fifo_preds[idx])
rl_pred = Image.open(rl_preds[idx]).convert("RGB")
fifo_pred = Image.open(fifo_preds[idx]).convert("RGB")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,12))
ax1.imshow(rl_pred)
ax2.imshow(fifo_pred)
ax1.tick_params(
    axis='both', which='both',
    bottom=False, top=False, left=False, right=False,
    labelbottom=False, labelleft=False
)
ax2.tick_params(
    axis='both', which='both',
    bottom=False, top=False, left=False, right=False,
    labelbottom=False, labelleft=False
)
```
