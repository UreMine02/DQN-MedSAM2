```python
import SimpleITK as sitk
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
slice_index = 90
```

```python
medsam2_ct_image_path = f"/mnt/fred/medical_image/btcv/Test/image/img0003/{slice_index}.jpg" 
medsam2_segmentation_label_path = f"/mnt/fred/medical_image/btcv/Test/mask/img0003/{slice_index}.npy"
medsam2_ct_image = np.array(Image.open(medsam2_ct_image_path).resize((1024, 1024)))
medsam2_segmentation_label = np.array(Image.fromarray(np.load(medsam2_segmentation_label_path)).resize((1024, 1024)))
```

```python
ct_image_path = "/mnt/fred/MedSeg/btcv_source/Abdomen/Abdomen/RawData/Training/img/img0003.nii.gz" 
segmentation_label_path = "/mnt/fred/MedSeg/btcv_source/Abdomen/Abdomen/RawData/Training/label/label0003.nii.gz"
ct_image = sitk.ReadImage(ct_image_path)
segmentation_label = sitk.ReadImage(segmentation_label_path)
```

```python
ct_image_array = sitk.GetArrayFromImage(ct_image)
segmentation_label_array = sitk.GetArrayFromImage(segmentation_label)
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(medsam2_ct_image)
ax[0].imshow(medsam2_segmentation_label, alpha=0.7)

ax[1].imshow(ct_image_array[slice_index, :, :])
ax[1].imshow(segmentation_label_array[slice_index, :, :], alpha=0.7)
ax[1].invert_yaxis()
```
